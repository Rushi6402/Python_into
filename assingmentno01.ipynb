```python
# Write a Python program that performs the following operations with lists and dictionaries:
# Map two lists into a dictionary.
# Convert a list of tuples into a dictionary.
# Clone or copy a list.
# Find the list of words that are longer than n from a given list of words.
# Remove duplicates from a dictionary.


list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
dict_mapped = dict(zip(list1, list2))
print("Mapped Dictionary:", dict_mapped)

lst_tuples = [('x', 10), ('y', 20), ('z', 30)]
dict_from_tuples = dict(lst_tuples)
dict_from_tuples

lst=[1,2,3,4,55,6]
copied_list=lst[:]
print("Copied List:", copied_list)  
print("original_list",lst)

# Find the list of words that are longer than n from a given list of words.
words = ["apple", "banana", "kiwi", "cherry", "mango"] 
 
n=int(input("Enter the number:"))
a=[]
for w in words:
    if  len(w)>n:
        a.append(w) 

print("Words longer than", n, ":", a)



# Remove duplicates from a dictionary.
dict_with_duplicates = {'a': 1, 'b': 2, 'c': 1, 'd': 3, 'e': 2}
seen_values = set()
dict_no_duplicates = {}
for key,value in dict_with_duplicates.items():
   if value not in seen_values:
       dict_no_duplicates[key]=value
       seen_values.add(value)

print("Dictionary after removing duplicates:", dict_no_duplicates)


    
```

```python
# Write a python code to accept a District and Print its State
district=input("Enter the district name: ")

district_state = {
    "Pune": "Maharashtra",
    "Mumbai": "Maharashtra",
    "Nashik": "Maharashtra",
    "Nagpur": "Maharashtra",
    "Bengaluru": "Karnataka",
    "Mysuru": "Karnataka",
    "Belagavi": "Karnataka",
    "Chennai": "Tamil Nadu",
    "Coimbatore": "Tamil Nadu",
    "Madurai": "Tamil Nadu",
    "Ahmedabad": "Gujarat",
    "Surat": "Gujarat",
    "Vadodara": "Gujarat"
}

if district in district_state:
    print(f"The state for district '{district}' is: {district_state[district]}")
else:
    print(f"The district '{district}' is not found in the list.")   
```
