Question 2: [10 Marks]
Write a program using reduce with a lambda function to find the product of all numbers in a list and
another program to sum the numbers and then divide by the length of the list to find the average.


```python
from functools import reduce

numbers=[2,3,4,56,7,8]

product=reduce(lambda x,y:x*y,numbers)

print(product)
```

```python
from functools import reduce

numbers=[20,304,50,59]

dived_by_length=reduce(lambda x,y:x/y,numbers)

print(dived_by_length)

print("to find average",dived_by_length/len(numbers))
```

Create the following dataset and do simple preprocessing
Dataset Name: Data.CSV which contains Country, Age, Salary, Purchased (in the form of Yes/No).
Write a program in python to perform following tasks:
Import Dataset.
Replace missing value of salary,age column with mean of that Column.
Data.csv have two categorical column (the country column, and the purchased column). Apply OneHot coding on Country
column.

```python

import pandas as pd
import numpy as np

# data={'name':['Alice','Bob','Charlie','David','Eva'],
#       'age':[24,30,22,35,28],
#       'city':['New York','Los Angeles','Chicago','Houston','Phoenix'], 
#       'salary':[2000,3000,   100 ,4000,],
#       'purchased':['yes','no','yes','no','yes']
# }
# df=pd.DataFrame(data)



data=pd.read_csv('data.csv')
data

# eplace missing value of salary,age column with mean of that Column.
data['salary']=data['salary'].fillna(data['salary'].mean())
data

data['age']=data['age'].fillna(data['age'].mean())


# Data.csv have two categorical column (the country column, and the purchased column). Apply OneHot coding on Country
# column.

country_encoded = pd.get_dummies(data['Country'])

data = pd.concat([country_encoded, data.drop('Country', axis=1)], axis=1)

data




```
